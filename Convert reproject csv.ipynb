We wanted to reproject the group 4's dataset to EPSG:3413. We used the ogr2ogr command to convert the csv file to geopackage. The source projection was kept as EPSG:4326 and target as EPSG:3413.



```python
%%bash
ogr2ogr -f GPKG output.gpkg Greenland_CDOM_2017_2018.csv -oo X_POSSIBLE_NAMES=LONGITUDE -oo Y_POSSIBLE_NAMES=LATITUDE -s_srs 'EPSG:4326' -t_srs 'EPSG:3413'
```
